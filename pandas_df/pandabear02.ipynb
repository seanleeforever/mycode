```python
# python3 -m pip install pandas
import pandas as pd
# This line is commented out. It is only necessary if using matplotlib
# at the CLI
# matplotlib.use('Agg')
# create some graphs
import matplotlib.pyplot as plt
```

```python
# define the name of our xls file
excel_file = '~/mycode/pandas_df/movies.xls'

# Choose the first column "Title" as
# index (index=0)
movies_sheet1 = pd.read_excel(excel_file, sheet_name=0, index_col=0)
# DF has 5 rows and 24 columns (indexed by title)
movies_sheet1.head()
```

```python
# grab the next 2 sheets as well
movies_sheet2 = pd.read_excel (excel_file, sheet_name=1, index_col=0)
#DF has 5 rows and 24 columns (index by title)
movies_sheet2.head()

```

```python
movies_sheet3 = pd.read_excel (excel_file, sheet_name=2, index_col=0)
#DF has 5 rows and 24 columns (index by title)
movies_sheet3.head()
```

```python
# combine all DFs into a single DF called movies
display (type (movies_sheet1))
movies = pd.concat([movies_sheet1,movies_sheet2,movies_sheet3])
# number of rows and columns (5042,24)
movies.shape
```

```python
# create a stacked bar graph
sorted_by_gross = movies.sort_values(["Gross Earnings"], ascending=False)
# Data is sorted by values in a column
# display the top 10 movies by Gross Earnings.
# passing the 10 values to head returns the top 10 not the default 5
sorted_by_gross.head(10)
```

```python
# create a stacked bar graph
sorted_by_gross["Gross Earnings"].head(10).plot(kind="barh")
plt.savefig("/home/student/static/stackedbar.png", bbox_inches='tight')
```

```python
sorted_by_gross_1900 = movies_sheet1.sort_values(["Gross Earnings"], ascending=False)
# create a stacked bar graph, and capture the object returned
ax = sorted_by_gross_1900['Gross Earnings'].head(10).plot(kind="barh", title="Highest Grossing Box Office Hits of 1900s")
ax.set_ylabel("Movie Titles")  # set the ylabel on the graph object
ax.set_xlabel("Cash Earnings per 100 million")  # set the xlabel on the graph object
plt.savefig("/home/student/static/stackedbar1900s.png", bbox_inches='tight')
```
