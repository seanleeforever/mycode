```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import json
import folium
from folium import plugins
```

```python
# read from LAcity.org JSON link and save as data frame
df=pd.read_json("https://data.lacity.org/resource/d5tf-ez2w.json")
```

```python
#Check the Data structure
df
```

```python
#get the geolocaiton information, which is under location1 attribuate as a dictionary.
#extract that dictionary and add onto the dataframe
df2=df.join(df.location_1.apply(pd.Series))
df2
```

```python
#only care about  geo locationd
noCareList = df2.columns
noCareList
```

```python
toDrop=['area','date_occ','time_occ','dr_no','date_rptd',
       'rpt_dist_no', 'crm_cd', 'crm_cd_desc', 'mocodes', 'vict_age',
       'vict_sex', 'vict_descent', 'premis_cd', 'premis_desc', 'location',
       'cross_street', 'location_1', ':@computed_region_qz3q_ghft',
       ':@computed_region_k96s_3jcv', ':@computed_region_tatf_ua23',
       ':@computed_region_ur2y_g4cx', ':@computed_region_kqwf_mjcx',
       ':@computed_region_2dna_qi2s',
       'human_address']
df2.drop(toDrop, inplace=True, axis=1)
df2
```

```python
#rename the column 
df2 = df2.rename(columns={'dr_no': 'Driver License Number', 'area_name':'Area Name', 'date_rptd': 'Date Reported', 'latitude': 'Latitude', 'longitude': 'Longitude'})
df2
```

```python
#get an idea of how many accidents happen at some location
accidents=df2.groupby(['Latitude','Longitude','Area Name']).size().reset_index(name='Count').sort_values(by='Count', ascending=False)
accidents.Count.max()
```

```python
#well. that is too specific, 4 decimal place is probabily too narrow, that's about 10 meters, let's increase that to 2 decimal
accidents.drop('Count',inplace=True, axis=1)
accidents[['Latitude','Longitude']]=accidents[['Latitude','Longitude']].astype(float).round(decimals=2)

ac=accidents.groupby(['Latitude','Longitude','Area Name']).size().reset_index(name='Count').sort_values(by='Count', ascending=False)
ac.Count.max()
```

```python
#give it some colors based on the number of accidents at that location
ac['Color'] = "grey"
ac['Color'] = ac['Count'].apply(lambda Count: "maroon" if Count >= 7 else
                              "firebrick" if Count >=5 and Count < 7  else
                              "red" if Count >=5 and Count < 5  else
                              "tomato" if Count >=3 and Count <4 else
                              "salmon" if Count >=2 and Count < 3 else
                              "mistyrose")
```

```python
#give it some size based on the number of accidents at that location
ac['Size']=ac['Count'].apply(lambda Count: 12 if Count >= 7 else
                              10 if Count >=5 and Count < 7  else
                              8 if Count >=5 and Count < 5  else
                              6 if Count >=3 and Count <4 else
                              4 if Count >=2 and Count < 3 else
                              2)
```

```python
ac
```

```python
ac_map=folium.Map([34.03438,-118.24256],zoom_start=11)
for lat,lon,area,color,count,size in zip(ac['Latitude'],ac['Longitude'],ac['Area Name'],ac['Color'],ac['Count'],ac['Size']):
     folium.CircleMarker([lat, lon],
                            popup='<h5>%s</h5>\n Number Accidents: %s' %(area,count),
                            radius=size,
                            color='b',
                            fill=True,
                            fill_opacity=0.7,
                            fill_color=color,
                           ).add_to(ac_map)
        
ac_map
```

```python
# convert to  format for heatmap, dont' reallly care about the small accidents 
ac2=ac.loc[ac['Count']>=2]
ac2.reset_index(drop = True)
```

```python
ac2_data = ac2[['Latitude', 'Longitude']].values

# plot heatmap
ac_map.add_child(plugins.HeatMap(ac2_data, radius=15))
ac_map
```

```python
from flask import Flask

app = Flask(__name__)

# HTTP GET /
# create a map that returns hershey 
@app.route('/')
def index():
    return ac_map._repr_html_()

if __name__ == '__main__':
    app.run(host="0.0.0.0", port=2224, debug=True)
```
